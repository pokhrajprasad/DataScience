```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Line Plot')

plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=2,markersize=9)
plt.grid(True)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Line Plot')

plt.show()
```

```python
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.figure(figsize=(9,5))

plt.subplot(2,2,1)
plt.plot(x,y1,color='green',linestyle='--',marker='o',linewidth=1,markersize=1)
plt.grid(True)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Line Plot')

plt.subplot(2,2,2)
plt.plot(x,y2,color='red',linestyle='--',marker='o',linewidth=1,markersize=1)
plt.grid(True)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Line Plot')

plt.subplot(2,2,3)
plt.plot(x,y1,color='blue',linestyle='--',marker='o',linewidth=1,markersize=1)
plt.grid(True)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Line Plot')

plt.subplot(2,2,4)
plt.plot(x,y2,color='black',linestyle='--',marker='o',linewidth=1,markersize=1)
plt.grid(True)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Line Plot')

plt.show()
```

```python
categories=['A','B','C','D','E']
values=[5,7,3,8,6]

plt.bar(categories,values,color='purple')
plt.title('Bar plot')
```
